#Create Multiple Bar Chart
#For each of the Legal Amazon state, compare the July Fire Count vs that of the combined July, August, September average count
#Conclusion: Most of the high fire counts are attributed to the higher 3mth season's fire count, compared to just July alone.
#Both Amapa and Roraima have considerably lower "seasonal" fire count than the standalone July's, compared to the rest of the states.

```python
w = 0.4

x = ["Acre", "Amapa","Pará","Amazonas","Rondonia","Tocantins","Roraima","Maranhao","Mata_Groso"]

Avg_Acre= Acre/20
Avg_Amapa= Amapa/20
Avg_Pará= Pará/20
Avg_Amazonas= Amazonas/20
Avg_Rondonia= Rondonia/20
Avg_Tocantins= Tocantins/20
Avg_Roraima= Roraima/20
Avg_Maranhao= Maranhao/20
Avg_Mata_Groso= Mata_Groso/20

state=[Avg_Acre, Avg_Amapa,Avg_Pará,Avg_Amazonas,Avg_Rondonia,Avg_Tocantins,Avg_Roraima,Avg_Maranhao,Avg_Mata_Groso]
Avg_Summer= [Avg_Summer_mths_Acre,Avg_Summer_mths_Amapa,Avg_Summer_mths_Pará,
             Avg_Summer_mths_Amazonas,Avg_Summer_mths_Rondonia,Avg_Summer_mths_Tocantin,
             Avg_Summer_mths_Roraima,Avg_Summer_mths_Maranhao,Avg_Summer_mths_Mato_Grosso]
```

```python
#Resize the graph
plt.figure(figsize=(10,3),dpi=250)

bar1 = np.arange(len(x))
bar2 = [ i +w for i in bar1]

plt.bar(bar1,state, w, label = "Julho")
plt.bar(bar2, Avg_Summer, w, label = "Avg 3 Months")

plt.xlabel("Legal Amazon States")
plt.ylabel("Fire Count")
plt.title("Julho vs Avg of Julho, Agosto, Setembro")

#Change the tick location and label from x
plt.xticks(bar1 + w/2, labels = x)
#plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

plt.show()
```
